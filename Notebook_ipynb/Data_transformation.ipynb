# Importing Necessary Libraries

```python
import pandas as pd
import numpy as np
import os
```

# Reading the Dataset

```python
df = pd.read_csv('/Users/rolex/Desktop/Dissertation/Data/nyc_taxi_data.csv')
```

# Convert 'tpep_pickup_datetime' and 'tpep_dropoff_datetime' columns to datetime format

```python
df = df.assign(
    tpep_pickup_datetime=pd.to_datetime(df['tpep_pickup_datetime']),
    tpep_dropoff_datetime=pd.to_datetime(df['tpep_dropoff_datetime'])
)
```

```python
df.info()
```

# Remove duplicate rows and reset index, keeping the new index as 'trip_id'

```python
df = df.drop_duplicates()
df = df.reset_index(drop=True)
df['trip_id'] = range(len(df))
```

```python
df.head()
```

# Select the datetime columns and reset the index

```python
datetime_dim = df.loc[:, ['tpep_pickup_datetime', 'tpep_dropoff_datetime']].reset_index(drop=True)
datetime_dim
```

# Create new columns based on 'tpep_pickup_datetime' in datetime_dim

```python
datetime_dim = datetime_dim.assign(
    pickup_hour=datetime_dim['tpep_pickup_datetime'].dt.hour,
    pickup_day=datetime_dim['tpep_pickup_datetime'].dt.day,
    pickup_month=datetime_dim['tpep_pickup_datetime'].dt.month,
    pickup_year=datetime_dim['tpep_pickup_datetime'].dt.year,
    pickup_weekday=datetime_dim['tpep_pickup_datetime'].dt.weekday
)
```

# Create new columns based on 'tpep_dropoff_datetime' in datetime_dim

```python
datetime_dim = datetime_dim.assign(
    drop_hour=datetime_dim['tpep_dropoff_datetime'].dt.hour,
    drop_day=datetime_dim['tpep_dropoff_datetime'].dt.day,
    drop_month=datetime_dim['tpep_dropoff_datetime'].dt.month,
    drop_year=datetime_dim['tpep_dropoff_datetime'].dt.year,
    drop_weekday=datetime_dim['tpep_dropoff_datetime'].dt.weekday
)
```

# Add 'datetime_id' as a column based on the DataFrame's index

```python
datetime_dim['datetime_id'] = datetime_dim.index

datetime_dim = datetime_dim[['datetime_id', 
                             'tpep_pickup_datetime', 'pickup_hour', 'pickup_day', 'pickup_month', 'pickup_year', 'pickup_weekday',
                             'tpep_dropoff_datetime', 'drop_hour', 'drop_day', 'drop_month', 'drop_year', 'drop_weekday']]

datetime_dim.head()
```

# New DataFrame with 'passenger_count' and add an 'id' column based on the index

```python
passenger_count_dim = df[['passenger_count']].reset_index()
passenger_count_dim = passenger_count_dim.rename(columns={'index': 'passenger_count_id'})

passenger_count_dim = passenger_count_dim[['passenger_count_id', 'passenger_count']]
passenger_count_dim.head()
```

# DataFrame with 'trip_distance' and include an 'id' column based on the index

```python
trip_distance_dim = df[['trip_distance']].reset_index().rename(columns={'index': 'trip_distance_id'})

trip_distance_dim = trip_distance_dim[['trip_distance_id', 'trip_distance']]
trip_distance_dim.head()
```

# mapping for rate codes

```python
rate_code_type = {
    1: "Standard rate",
    2: "JFK",
    3: "Newark",
    4: "Nassau or Westchester",
    5: "Negotiated fare",
    6: "Group ride"
}

# new DataFrame with 'RatecodeID' and reset index
rate_code_dim = df[['RatecodeID']].reset_index()
rate_code_dim = rate_code_dim.rename(columns={'index': 'rate_code_id'})

rate_code_dim['rate_code_name'] = rate_code_dim['RatecodeID'].map(rate_code_type)

rate_code_dim = rate_code_dim[['rate_code_id', 'RatecodeID', 'rate_code_name']]
rate_code_dim.head()
```

# Create a DataFrame with 'pickup_longitude' and 'pickup_latitude', and reset the index

```python
pickup_location_dim = df[['pickup_longitude', 'pickup_latitude']].reset_index()

pickup_location_dim = pickup_location_dim.rename(columns={'index': 'pickup_location_id'})

pickup_location_dim = pickup_location_dim[['pickup_location_id', 'pickup_latitude', 'pickup_longitude']]
pickup_location_dim.head()
```

# Select 'dropoff_longitude' and 'dropoff_latitude' columns, and reset the index

```python
dropoff_location_dim = df[['dropoff_longitude', 'dropoff_latitude']].reset_index()

dropoff_location_dim.rename(columns={'index': 'dropoff_location_id'}, inplace=True)

dropoff_location_dim = dropoff_location_dim[['dropoff_location_id', 'dropoff_latitude', 'dropoff_longitude']]
dropoff_location_dim.head()
```

# mapping for payment types

```python
payment_type_name = {
    1: "Credit card",
    2: "Cash",
    3: "No charge",
    4: "Dispute",
    5: "Unknown",
    6: "Voided trip"
}

# Create a DataFrame with 'payment_type' and reset the index
payment_type_dim = df[['payment_type']].reset_index()
payment_type_dim = payment_type_dim.rename(columns={'index': 'payment_type_id'})

payment_type_dim['payment_type_name'] = payment_type_dim['payment_type'].map(payment_type_name)

payment_type_dim = payment_type_dim[['payment_type_id', 'payment_type', 'payment_type_name']]
payment_type_dim.tail()
```

# series of merges to combine the DataFrames

```python
fact_table = df \
    .merge(passenger_count_dim, how='left', left_on='trip_id', right_on='passenger_count_id') \
    .merge(trip_distance_dim, how='left', left_on='trip_id', right_on='trip_distance_id') \
    .merge(rate_code_dim, how='left', left_on='trip_id', right_on='rate_code_id') \
    .merge(pickup_location_dim, how='left', left_on='trip_id', right_on='pickup_location_id') \
    .merge(dropoff_location_dim, how='left', left_on='trip_id', right_on='dropoff_location_id') \
    .merge(datetime_dim, how='left', left_on='trip_id', right_on='datetime_id') \
    .merge(payment_type_dim, how='left', left_on='trip_id', right_on='payment_type_id')

fact_table = fact_table[['trip_id', 'VendorID', 'datetime_id', 'passenger_count_id',
                         'trip_distance_id', 'rate_code_id', 'store_and_fwd_flag', 'pickup_location_id', 
                         'dropoff_location_id', 'payment_type_id', 'fare_amount', 'extra', 
                         'mta_tax', 'tip_amount', 'tolls_amount', 'improvement_surcharge', 
                         'total_amount']]
```
